## Get the data

```python
import pandas as pd

df = pd.read_csv("../cleaned_dataset.csv")
df
```

```python
from sklearn.model_selection import train_test_split

tem_train, tem_test = train_test_split(df, test_size=0.2, random_state=42, shuffle=False)
```

## Modeling

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Fit the ARIMA model
arima_model = ARIMA(tem_train['Rainfall'], order=(1,0,0))  # ARIMA(p,d,q)
arima_model_fit = arima_model.fit()

# Make predictions
arima_pred_m = arima_model_fit.forecast(steps=len(tem_test))

# Calculate error
error = mean_squared_error(tem_test['Rainfall'], arima_pred_m)
print('Test MSE: %.3f' % error)

# Plot the results
plt.plot(tem_test.index, tem_test['Rainfall'], label='Actual')
plt.plot(tem_test.index, arima_pred_m, color='red', label='Forecast')
plt.xlabel('Date')
plt.ylabel('Difference in Rainfall')
plt.title('ARIMA Forecast')
plt.legend()
plt.show()
```
